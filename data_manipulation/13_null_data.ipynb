### Null data

```
import pandas as pd
import numpy as np
```

```
dict = {'Col1':[1,2,3,np.nan],
'Col2':[4, np.nan,6,7],
'Col3':['a','b','c', None]}
```

```
dict
```

```
df = pd.DataFrame(dict)
```

### Identify null values in a DataFrame

```
df.isnull()
```

### Identify null values with a numeric value 

```
df.isnull()*1
```

### Replace null values by a string

```
df.fillna('Missing')
```

### Replace null values by a statistical measure made with the column values

```
df.fillna(df.mean())
```

### Replace null values by interpolation values

```
df.interpolate()
```

### Eliminate null values

```
df.dropna()
```
