### LOC

```
df_books.loc[0:4, ['Name','Author']]
```

```
df_books.loc[:,['Reviews']] * -1
```

```
df_books.loc[:,['Author']] == 'JJ Smith'
```

To check whether any element in a DataFrame matches a comparison value, you can use the .any() method to collapse the boolean DataFrame into a single boolean value.

```
mask = (df_books.loc[:, ['Author']] == 'Stephen King').any(axis=1)

df_filtered = df_books.loc[mask, ['Name', 'Author', 'Year']]

df_filtered
```
