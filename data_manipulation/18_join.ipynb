### Join

```
import pandas as pd
import numpy as np
```

```
izq = pd.DataFrame({'A': ['A0','A1','A2'],
                    'B':['B0','B1','B2']},
                    index=['k0','k1','k2'])

der =pd.DataFrame({'C': ['C0','C1','C2'],
                   'D':['D0','D1','D2']},
                    index=['k0','k2','k3']) 
```

join() method is used to join two dataframes based on their indexes or columns. It will join the two dataframes using their index values by default. This is equivalent to calling izq.join(der, how='left'), which performs a left join of izq with der.

```
izq.join(der)
```

 izq.join(der, how='outer') statement is used to perform an outer join between two pandas dataframes izq and der. An outer join returns all rows from both dataframes, with NaN values filled in for missing values.

```
izq.join(der, how = 'outer')
```
