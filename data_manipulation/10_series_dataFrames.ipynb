```
import pandas as pd
```

### Series

```
psg_players = pd.Series(['Navas','Mbappe','Neymar','Messi'],
    index=[1,7,10,30]
    )
psg_players
```

Using dictionaries

```
psd_dict = {1: 'Navas', 7: 'Mbappe', 10: 'Neymar', 30: 'Messi'}
pd.Series(psd_dict)
```

```
psg_players[1]
```

Access values by integer location

```
psg_players.iloc[0:3]
```

### Data Frames

```
dict = {'jugador': ['Navas','Mbappe','Neymar','Messi'],
'height': [180.0, 172.0, 169.6, 160.7],
'goals': [10, 20, 30, 40]}

pd.DataFrame(dict, index=[1,7,10,30])
```
