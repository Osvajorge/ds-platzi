### Filtering by conditions

```
import pandas as pd
```

```
df_books = pd.read_csv('/work/bestsellers.csv', sep= ',', header=0)
df_books.head(2)
```

```
greater_than_2016 = df_books['Year'] > 2016
```

```
fiction_genre = df_books['Genre'] == 'Fiction'
```

```
df_books[fiction_genre & greater_than_2016]
```

### Negating the condition

```
df_books[~greater_than_2016]
```
