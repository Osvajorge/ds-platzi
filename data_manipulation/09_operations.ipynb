```
import numpy as np
```

```
my_list = [1,2]
my_list
```

Vanilla python 

```
my_list * 2
```

Using numpy

```
second_list = np.arange(0,10)
print(second_list * 2)
print(second_list + 2)
print(second_list / 2)
print(second_list ** 2)
```

```
matrix = second_list.reshape(2,5)
matrix2 = matrix.copy()
```

```
matrix + matrix2
```

```
np.matmul(matrix,matrix2.T)
```

```
matrix @ matrix2.T
```
