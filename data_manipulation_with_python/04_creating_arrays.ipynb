### Create a Python list containing the numbers from 0 to 9 using the built-in `range` function

```
import numpy as np
l1 = list(range(0,10))
print(l1, type(l1))
```

Create a NumPy array containing the numbers from 0 to 9 using the `np.arange` function

```
l2 = np.arange(0,10)
print(l2, type(l2))
```

Create a NumPy array containing the even numbers from 0 to 20 using the `np.arange` function

```
np.arange(0,22,2)
```

Create a NumPy array of zeros with shape (3, 3) using the `np.zeros` function

```
np.zeros((3,3))
```

Create a NumPy array of ones with shape (5, 5) using the `np.ones` function

```
np.ones((5,5))
```

Create a NumPy array of 10 equally spaced numbers between 0 and 10 (inclusive) using the `np.linspace` function

```
np.linspace(0,10,10)
```

Create a 4x4 identity matrix using the `np.eye` function

```
np.eye(4)
```

Create a 1D NumPy array of 4 random numbers between 0 and 1 using the `np.random.rand` function

```
np.random.rand(4)
```

Create a 2D NumPy array of shape (4, 4) with random numbers between 0 and 1 using the `np.random.rand` function

```
np.random.rand(4,4)
```

Generate a random integer between 1 and 15 (inclusive) using the `np.random.randint` function

```
np.random.randint(1,15)
```
