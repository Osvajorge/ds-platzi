# Functions

```
import numpy as np
```

```
arr = np.random.randint(1,20,10)
arr
```

```
matrix = arr.reshape(2,5)
matrix
```

### max() and min()

```
arr.max()
```

```
print(matrix.max())
#max number of axis one
print(matrix.max(1))
#max number of axis zero
print(matrix.max(0))
#index of the max number, if there are more than one, it selects the lower
print(matrix.argmax(0))
```

```
print(matrix.min())
#min number of axis one
print(matrix.min(1))
#min number of axis zero
print(matrix.min(0))
#index of the min number, if there are more than one, it selects the lower
print(matrix.argmin(0))
```

### PTP (peak to peak)

```
#Difference betwen the greater and lower value
arr.ptp()
#Difference betwen the greater and lower value in 0 axis
arr.ptp(0)
```

### Percentile, sort, median, mean  variance and std

```
#sorted array
print(np.sort(arr))
#Higher
print(np.percentile(arr,100))
#median
print(np.percentile(arr,50))
print(np.median(arr))
#median matrix
print(np.median(matrix,1))
#stanrdard deviation
print(np.std(arr))
#variance
print(np.var(arr))
#mean
print(np.mean(arr))
```

### Concatenate

```
a = np.array([[1,2],[3,4]])
print(a, "\n")
b = np.array([5,6])
print(b,"\n")
b = np.expand_dims(b, axis=0)
print(b,"\n")
print(np.concatenate((a,b), axis=0), "\n")
print(np.concatenate((a,b.T), axis=1))
```
