# Data types

```
import numpy as np
```

int64

```
arr = np.array([1,2,3,4])
arr.dtype
```

int64 to float64

```
arr = np.array([1,2,3,4], dtype='float64')
arr.dtype
arr
```

int64 to float64 using numpy astype

```
second_array = np.array([1,2,3,4,5])
second_array
second_array.astype(np.float64)
```

int64 to bool_

```
third_array = np.array([0,1,2,3,4,5])
third_array.astype(np.bool_)
```

int64 to string_

```
fourth_array = np.array([0,1,2,3,4,5])
fourth_array.astype(np.string_)
```

string to int8

```
fifth_array = np.array(['0','1','2',])
fifth_array.astype(np.int8)
```
