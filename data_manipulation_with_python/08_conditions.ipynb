## Conditions

```
import numpy as np
```

```
arr = np.linspace(1,10,10, dtype='int8')
arr
```

```
ind_contitions = arr > 5
ind_contitions
```

Array condition

```
arr[ind_contitions] #arr[arr > 5]
```

Multiple conditions 

```
arr[(arr > 5) & (arr<9)]
```

Modify an array through a condition

```
arr[arr >5] = 12
arr
```
