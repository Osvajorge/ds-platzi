### Shape and Reshape

```
import numpy as np
```

Generates an array with random integers

```
arr = np.random.randint(1,10,(3,2))
print(arr)
arr.shape
```

Reshape the original array 

```
arr.reshape(1,6)
```

```
arr.reshape(2,3)
```

```
np.reshape(arr,(6,1))
```

Reshape as 'C' language 

```
np.reshape(arr,(2,3),'C')
```

Reshape as 'Fortran' language 

```
np.reshape(arr,(2,3),'F')
```

Reshape as randomly, optimized as memory does it 

```
np.reshape(arr,(2,3),'A')
```
