### Bar plot

```
import numpy as np 
import matplotlib.pyplot as plt
```

```
countries = ['India', 'Japan', 'Mexico', 'Colombia', 'Germany']
population = [500, 800, 900, 1000, 300]
```

plt.bar() is used to create bar charts or bar graphs, which are used to display and compare the values of different categories.

The width parameter sets the width of the bars, and the bottom parameter sets the position of the baseline of the bars. The align parameter sets the alignment of the bars with the x-coordinates, and the data parameter is used to specify the data source for the plot.

```
plt.bar(countries,population);
```

```
plt.bar(countries,population,  width=0.5, color=["aqua", "grey", "teal", "crimson", "violet"])
plt.show()
```

plt.xticks() set the locations and labels of the tick marks on the x-axis of a plot. np.arange(5) is a NumPy array that creates a sequence of numbers from 0 to 4, which will be used as the tick locations on the x-axis. The tick labels are specified by the tuple ('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'). The rotation parameter rotates the tick labels by the specified angle (in degrees), which in this case is 45 degrees.

```
plt.bar(countries,population, width=0.5, color=["aqua", "grey", "teal", "crimson", "violet"])
plt.xticks(np.arange(5), ('India','Japon', 'Mexico', 'Colombia', 'Alemania'), rotation = 45)
plt.title('Countries pouplation')
plt.ylabel('Population')
plt.xlabel('Country');
```

plt.barh() is a function in the Matplotlib library in Python, which is used to create horizontal bar charts or bar graphs. A horizontal bar chart displays the data on the y-axis instead of the x-axis, making it useful when you want to compare the values of different categories.

```
plt.barh(countries,population)
plt.title('Countries pouplation')
plt.xlabel('Population')
plt.ylabel('Country');
```
