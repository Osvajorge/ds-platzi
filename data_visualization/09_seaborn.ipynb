### Seaborn

Seaborn is a Python data visualization library based on Matplotlib, which allows you to create informative and visually appealing statistical graphics. It provides a high-level interface for creating data visualizations with a few lines of code, making it easy to create complex plots like heatmaps, scatter plots, bar plots, and many others.

```
import seaborn as sns
import matplotlib.pyplot as plt
```

### Set

In Seaborn, a set is a collection of colors that are designed to work together for visualizing categorical data. Seaborn provides several built-in sets, each of which contains a different number of colors.

```
sns.set(style='dark', palette='icefire', font='DejaVu Sans', font_scale=1)
sns.barplot(x=['A','B','C'], y=[1,2,3])
plt.show();
```
