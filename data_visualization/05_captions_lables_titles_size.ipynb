### Captions, labels, titles, size

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,5,11)
y  = x ** 2
```

Creates 2 charts

```
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'g')
ax2.plot(y,x, 'r')
plt.show()
```

set_title() set the title of a plot or subplot

```
fig, (ax1, ax2) = plt.subplots(1,2)

ax1.set_title('X - Y Relationship')
ax1.plot(x,y, 'g')
ax2.set_title('Y - X Relationship')
ax2.plot(y,x, 'r')
plt.show()
```

set_label() set the label of a plot or subplot.

```
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b', )
ax1.set_title('X - Y Relationship')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

ax2.plot(y,x, 'r')
ax2.set_title('Y - X Relationship')
ax2.set_xlabel("Y")
ax2.set_ylabel("X")

plt.show()
```

The figsize argument specifies the width and height of the figure in inches. In this case, the figure will be 1 inch wide and 10 inches tall, which is an unusual aspect ratio that will make the subplots very elongated vertically.

```
fig, (ax1,ax2) = plt.subplots(1,2, figsize= (7,6))
```

The label parameter is being used to specify the label of the line that will appear in the legend. In this case, the label is set to "sin(x)".

```
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b', label = "sin(x)")
ax1.set_title('X - Y Relationship')
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.legend();
```
