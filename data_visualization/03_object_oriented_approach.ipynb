### Object oriented approach

In this method, a figure object is created first, which can contain one or more subplots. Then, plot objects are created within each subplot to generate the desired visualization. These plot objects can be modified and customized using various methods and properties.

The object-oriented approach allows for greater control and flexibility in creating and modifying plots, compared to the simpler pyplot interface. It also enables more complex and sophisticated visualizations to be created, by allowing for the composition of multiple plot types and elements.

```
import matplotlib.pyplot as plt
import numpy as np
```

```
x = np.linspace(0,5,11)
y = x ** 2
```

fig = plt.figure() creates a new Figure object, which is the top-level container for all the plot elements.

axes = fig.add_axes([0.1,0.1,0.5,0.9]) creates a new Axes object within the Figure. The [0.1,0.1,0.5,0.9] argument specifies the position and size of the Axes within the Figure, as a list of four values [left, bottom, width, height]. In this case, the Axes is positioned at 10% from the left and bottom edges of the Figure, and has a width of 50% and a height of 90%.

axes.plot(x,y, 'b') plots the data specified by the arrays x and y on the Axes using a blue line ('b'). This line style argument specifies the color and style of the line.

```
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y, 'b');
```

The figure contains two subplots, where the first subplot (axes) is plotted with x and y data using a blue line, and the second subplot (axes2) is plotted with y and x data using a red line. Additionally, the background color of the second subplot has been set to grey using the set_facecolor method.

```
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.17,0.55,0.4,0.3])

axes.plot(x,y, 'b')
axes2.plot(y,x, 'r')
axes2.set_facecolor('grey')
plt.show()
```
