### Heatmap

```
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips.head()
```

```
tips.corr()
```

```
sns.heatmap(tips.corr());
```

```
sns.heatmap(tips.corr(), annot=True, 
    cmap='icefire', linewidths=1, linecolor='black', vmin=0.5, vmax=1, cbar=True);
```
