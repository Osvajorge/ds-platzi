### Subplots

In Matplotlib, subplots refer to the capability to create multiple plots within a single figure. Subplots allow you to organize and visualize multiple data sets or different aspects of a single data set in a clear and concise manner.

To create subplots in Matplotlib, you can use the plt.subplots() function. This function returns a figure object and an array of axes objects. The number of rows and columns of the subplots can be specified as arguments to the plt.subplots() function.

```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.linspace(0,5,11)
y = x ** 2
```

fig, axes = plt.subplots() creates a figure object and one or more subplot objects within it, and returns them as a tuple. The first element of the tuple is the figure object, which is stored in the variable fig, and the second element is the subplot object, which is stored in the variable axes.

```
fig, axes = plt.subplots()
axes.plot(x,y);
```

fig, axes = plt.subplots(nrows=1, ncols=2)  creates a figure object and two subplot objects within it, arranged side by side. The nrows parameter specifies the number of rows of subplots, and the ncols parameter specifies the number of columns of subplots. In this case, nrows=1 and ncols=2 creates a figure with one row and two columns of subplots. The fig variable is assigned to the figure object, and the axes variable is assigned to an array of the two subplot objects.

```
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y, 'b')
axes[1].plot(x,y, 'r')
plt.show()
```

```
fig, (axes1,axes2) = plt.subplots(nrows=1, ncols=2)
axes1.plot(x,y, 'b')
axes2.plot(x,y, 'r')
plt.show()
```

plt.subplots(2,4) creates a figure with 2 rows and 4 columns of subplots. The function call creates a figure with 8 subplots arranged in a 2x4 grid

In this case, two subplots are created, one in the top-left corner (axes[0,0]) and another in the top-right corner (axes[0,1]). 

The fig.tight_layout() function is then called to adjust the spacing between subplots so that the axes labels and titles do not overlap.

```
fig, axes = plt.subplots(2,4)
axes[0,0].plot(x,y)
axes[0,1].plot(y,x, 'r')

fig.tight_layout()
```

```
def eq(x, y):
    return 0.2*x + 0.3*y + 0.4 + 0.2*(x - x[0]*0.2 + y[0])
```

```
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2,4)
ax1.plot(x, np.sin(x), 'b')
ax2.plot(y, np.cos(x), 'purple')
ax3.plot(x, np.tan(y), 'orange')
ax4.plot(y, np.cos(y)**2, 'black')
ax5.plot(x, (10*x) + y)
ax6.plot(y, (-10*y) + x, 'mD:')
ax7.plot(x - y, 'teal')
ax8.plot(x, eq(x, y), 'r')
fig.tight_layout()
```
