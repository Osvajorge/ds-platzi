### Plot Types

```
import numpy as np
import matplotlib.pyplot as plt
import random
```

```
data = np.random.randint(1,50,100)
```

A histogram is a graphical representation of the distribution of a dataset, where the data is divided into a set of intervals (called "bins") and the count of data points that fall into each bin is plotted.

The plt.hist function takes as input the data to be plotted, and optionally several other arguments that control the appearance of the plot, such as the number of bins to use, the range of values to include, and the color of the bars.

```
plt.hist(data, color='#0A4D68');
```

The bins parameter is set to 50, which means the data will be divided into 50 equally-spaced intervals

```
plt.hist(data, bins=50, color='#05BFDB')
plt.show()
```

histtype is a parameter that specifies the type of histogram to be drawn. It can take several values, including:

    bar (default): This draws a standard histogram with vertical bars for each bin.
    barstacked: This draws a stacked histogram, where the bars for each bin are stacked on top of each other.
    step: This draws a histogram with the tops of the bars connected by a line, resulting in a "step" pattern.
    stepfilled: This is similar to the step type, but the area between the line and the x-axis is filled with color.

```
plt.hist(data, bins=10, histtype='step', color='#05BFDB') 
plt.show()
```

A boxplot is a graphical representation of the distribution of a dataset through five summary statistics: the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. The box in the boxplot represents the interquartile range (IQR), which is the distance between the first and third quartiles.

The boxplot variable is a dictionary-like object that contains several elements, including the medians, boxes, whiskers, and outliers of the boxplot.

```
# Without customization
#plt.boxplot(data)  
#plt.show()


boxplot = plt.boxplot(data)

# customize the median line color
median_line_color = '#00FFCA'

plt.setp(boxplot['medians'], color=median_line_color);
```

    vert=False: This argument sets the orientation of the boxplot to horizontal, so that the boxes and whiskers are drawn horizontally instead of vertically.
    patch_artist=True: This argument fills the boxes and whiskers with color instead of using black outlines and white fill.
    notch=True: This argument creates a notch around the median line to indicate the confidence interval of the median.

```
# Without customization
#plt.boxplot(data, vert=False, patch_artist=True, notch=True)
#plt.show()


boxplot = plt.boxplot(data, vert=False, patch_artist=True, notch=True)

# customize the median line color
median_line_color = '#00FFCA'

# fill the boxes with a color
box_color = '#088395'
for patch in boxplot['boxes']:
    patch.set_facecolor(box_color)


plt.setp(boxplot['medians'], color=median_line_color);
```

 showfliers=False, the outlier points will not be displayed in the plot.

For loop to iterate through each box in the boxplot (i.e., the rectangles that represent the interquartile range or IQR). For each box, the set_facecolor() method is used to set the color of the box to the box_color variable.

```
#data = np.append(data, 200)
#plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)  
#plt.show()

data = np.append(data, 200)
boxplot = plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)

# customize the median line color
median_line_color = '#FFDCB6'

# fill the boxes with a color
box_color = '#C9A7EB'
for patch in boxplot['boxes']:
    patch.set_facecolor(box_color)


plt.setp(boxplot['medians'], color=median_line_color);

```

 N = 50: This sets the number of points in the scatter plot to 50.
 x = np.random.rand(N): This generates an array of 50 random values between 0 and 1, which will be used as the x-coordinates of the points in the scatter plot.
y = np.random.rand(N): This generates an array of 50 random values between 0 and 1, which will be used as the y-coordinates of the points in the scatter plot.
area = (30 * np.random.rand(N)) **2: This generates an array of 50 random values between 0 and 900, which will be used as the areas of the points in the scatter plot. The formula (30 * np.random.rand(N)) **2 generates values between 0 and 30, multiplies them by 30, and then squares them to obtain areas between 0 and 900.
colors = np.random.rand(N): This generates an array of 50 random values between 0 and 1, which will be used as the colors of the points in the scatter plot.

```
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) **2
colors = np.random.rand(N)
```

plt.scatter(x, y, s=area): This creates a scatter plot with the x coordinates on the x-axis, the y coordinates on the y-axis, and the point sizes specified by the area array. The s parameter controls the size of the points. By default, the c parameter is not set, so the color of the points will be determined by the default color map.

```
plt.scatter(x,y, s=area)
plt.show()
```

```
plt.scatter(x,y, s=area, c= colors, marker = 'v') # change maker
plt.show()
```

```
plt.scatter(x,y, s=area, c= colors, marker = 'o', alpha= 0.5) # change transparency
plt.show()
```
