### Subplot

A subplot is a grid of plots that are displayed within a single figure. A subplot allows you to create multiple plots within a single figure and arrange them in a grid format. Each subplot can display different data or variations of the same data.

```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.linspace(0,5,11)
y = x ** 2
```

plt.subplot(1, 2, 1) creates a subplot in a 1x2 grid of subplots, with the current subplot being the first one (i.e., leftmost) in the grid.

In the example plt.subplot(1, 2, 1), 1 indicates that there is only one row in the subplot grid, 2 indicates that there are two columns in the subplot grid, and 1 indicates that the current subplot is the first (leftmost) one in the grid.

```
plt.subplot(1,2,1) #Chart 1
plt.plot(x,y,'r--')

plt.subplot(1,2,1) #Chart 2
plt.pie(y);
```

This will display a figure with two subplots side-by-side, with a line chart of y against x with a red dashed line style and a line chart of x against y with a blue dotted line style on the first subplot, and a pie chart of y on the second subplot.

```
plt.subplot(1,2,1) 
plt.plot(x,y, 'r--') # plot 1 
plt.plot(y,x, 'b:') #plot 2

plt.subplot(1,2,2)
plt.pie(y);
```

The first plt.subplot(2,1,1) creates the first subplot in a 2x1 grid of subplots, with the current subplot being the top one. Then, plt.plot(x,y,'r--') and plt.plot(x,y,'b:') plot two line charts of y against x with a red dashed line style and a blue dotted line style, respectively, on the first subplot.

The second plt.subplot(2,1,2) creates the second subplot in the same 2x1 grid, with the current subplot being the bottom one. Then, plt.pie(y) creates a pie chart of y on the second subplot.

```
plt.subplot(2,1,1) # chart 1
plt.plot(x,y,'r--') # plot 1
plt.plot(x,y,'b:') # plot 2

plt.subplot(2,1,2) #chart 2
plt.pie(y); # plot 1
```
