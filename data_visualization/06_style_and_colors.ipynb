### Style and colors

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,5,11)
y  = x ** 2
```

plt.style.available command returns a list of available plot styles that you can use in Matplotlib

```
print(plt.style.available)
```

plt.style.use('seaborn-whitegrid') sets the current plot style in Matplotlib to "seaborn-whitegrid".

```
plt.style.use('seaborn-whitegrid')
```

```
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+1, 'r--')
ax.plot(x,x+2, 'bo-')
ax.plot(x,x+3, 'g.:')
ax.plot(x,x+4, 'purple');
```

```
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+1,color = '#D426C8') #RGB
ax.plot(x,x+2,color = '#66FF89')

ax.plot(x,x+3,color = 'blue') #Common name
ax.plot(x,x+4, color = 'black');
```

alpha is a parameter that control the transparency of the plotted data. The alpha parameter ranges from 0 to 1, where 0 is completely transparent (invisible), and 1 is completely opaque (solid).

linewidth is a parameter that controls the width of the lines that are plotted in a figure.

```
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+1,color = '#D426C8', alpha= 0.5, linewidth= 18)
ax.plot(x,x+2,color = '#66FF89', linewidth= 3)
ax.plot(x,x+3,color = 'blue', linewidth= 5)
ax.plot(x,x+4, color = 'black', alpha = 0.3, linewidth= 12);
```

linestyle is a parameter that controls the style of the lines that are plotted in a figure. Some of the most commonly used line styles include solid lines ('-'), dashed lines ('--'), dotted lines (':'), and dash-dot lines ('-.').

marker is a parameter that controls the shape of the markers that are plotted in a figure.  Some of the most commonly used marker shapes include circles ('o'), squares ('s'), triangles ('^'), and diamonds ('d').

```
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+3,color = 'blue', linestyle = 'dashed', marker = 'x')
ax.plot(x,x+4, color = 'black',linestyle = '--', marker = 'P');
```

markersize is a parameter  that controls the size of the markers that are plotted in a figure.

markerfacecolor is a parameter that controls the fill color of the markers that are plotted in a figure.

```
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+3,color = 'blue', linestyle = 'dashed', marker = '8', markersize = 10, markerfacecolor= "#37D842")

ax.plot(x,x+4, color = 'black',linestyle = '--', marker = 'P', markerfacecolor="#FF0000");
```

Hack to check all the styles 

```
for i in plt.style.available:
  plt.style.use(i)
  fig, ax = plt.subplots(figsize=(2,2))
  ax.set_title(i, size=8)
plt.show()

```
