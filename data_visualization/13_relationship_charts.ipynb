### Relationship charts

```
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips.head(5)
```

Scatter plot using Seaborn

```
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day');
```

style parameter is used to determine the marker style for each point, based on the time column.

Size parameter determines the size of each point, based on the size column.

plt.legend() function is called to create a legend for the plot, with the loc parameter specifying the location of the legend and the bbox_to_anchor parameter specifying the anchor point of the legend.

```
sns.scatterplot(data= tips, x= 'total_bill', y ='tip', hue='day', style='time', size='size');
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5));
```

Adds a dictionary of marker styles to the markers parameter. The dictionary maps each value of the time column to a specific marker style. Specifically, D is used to represent Lunch and s is used to represent Dinner.

```
markers = {"Lunch": 'D', "Dinner": 's'}
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers);
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5));
```

sns.lineplot() creates line plots of data.

```
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size');
```

sns.relplot() is used to create various types of plots with added customization options. It is built on top of the FacetGrid class of the matplotlib library, which allows for creating subplots based on one or more variables.

```
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', kind='line');
```

Col parameter is used to create separate subplots based on the time column. This means that there will be two separate plots, one for Lunch and one for Dinner.

```
sns.relplot(data=tips, x='total_bill', y='tip',hue='day', style='time', size='size', kind='scatter', col='time');
```
